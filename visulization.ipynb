```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
 def process_privacy_data(filename):
    file = open(filename)
    dataMat = []
    for line in file.readlines():
        dataMat.append(float(line))
    return dataMat
```

```python
ClusterFed_user30_HR = process_privacy_data(r"./2021-09-12-15-28-43/_balance_True_cluster_6_user_5_ServerEpochs_60_ClusterEpochs_1_/logs/server/total_test_HR/total_test_HR.txt")
ClusterFed_user20_HR = process_privacy_data(r"./2021-09-12-15-28-43/_balance_True_cluster_4_user_5_ServerEpochs_60_ClusterEpochs_1_/logs/server/total_test_HR/total_test_HR.txt")
ClusterFed_user10_HR = process_privacy_data(r"./2021-09-12-15-28-43/_balance_True_cluster_2_user_5_ServerEpochs_60_ClusterEpochs_1_/logs/server/total_test_HR/total_test_HR.txt")

epochs = [x for x in range(0, 60, 1)]

plt.plot(epochs, ClusterFed_user10_HR, 'b'+'--', label='ClusterFed cluster 2 user 5',linewidth=2)
plt.plot(epochs, ClusterFed_user20_HR, 'm'+'--', label='ClusterFed cluster 4 user 5',linewidth=2)
plt.plot(epochs, ClusterFed_user30_HR, 'g'+'--', label='ClusterFed cluster 6 user 5',linewidth=2)


font1 = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 30,
}

plt.tick_params(labelsize=30)
plt.legend(prop=font1)
plt.xlabel('epochs', font1)
plt.ylabel('HR', font1)

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(16, 14)
plt.savefig(r"./result/ClusterFed_HR")
```
